```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data_frame = pd.read_csv("placement.csv")
print(data_frame.head())
```

```python
print(data_frame['cgpa'].head())
```

```python
plt.scatter(data_frame['cgpa'], data_frame['package'])
plt.xlabel('CGPA')
plt.ylabel('Package in LPA')
```

```python
X = data_frame.iloc[:,0:1]
print(X)
```

```python
Y = data_frame.iloc[:,-1]
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
```

```python
linear_regression.fit(X_train, Y_train)
```

```python
print(X_test.iloc[:10], end="\n*************\n")
print(Y_test.iloc[:10])
```

```python
# linear_regression.predict()
print(X_test.iloc[0].values.reshape(1, 1))
```

```python
print(linear_regression.predict(X_test))
```

```python
plt.scatter(X_train, Y_train)
plt.plot(X_train, linear_regression.predict(X_train), color="violet")
```

```python
theta_slope = linear_regression.coef_
print(f"The slope of the best-fit line is {theta_slope}")

intercept = linear_regression.intercept_
print(f"The intercept of the best-fit line is {intercept}")
```

```python
# Let's calculate for the cgpa 8.58

# By linear_regression ::
print(linear_regression.predict(X_test.iloc[0].values.reshape(1, 1)))

# By best-fit line ::
print((theta_slope * X_test.iloc[0]) + intercept)
```
